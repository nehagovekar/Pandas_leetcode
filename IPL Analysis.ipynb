# Pandas Part 5 - Filtering dataframe on a condition | Masking using boolean series

```python
#fetching rows depending on a given condition

mask = data['city']== 'Hyderabad'
mask
```

```python
data[mask]
```

```python
def get_city_match_count(city):
    mask= data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Rajkot')
```

```python
mask1 = data['city']=='Hyderabad'
mask2 = data['date']>'2010-01-01'
```

```python
data[mask1 & mask2] #when boolean series then use bitwise operator
```
