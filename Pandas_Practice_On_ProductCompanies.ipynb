```python
"""
Title: Amazon Product Reviews

Instruction: Given the tables provided below, solve the problem sets involving Pandas DataFrames.

Hint: Time yourself. Aim to solve each question spending no more than 7 to 8 minutes per each question. In addition, practice explaining your solution out loud.

### DataFrames

# df_product_sales
+------------+------------+------------+------------+--------+
| product_id | sale_date  | units_sold | sale_price | region |
+------------+------------+------------+------------+--------+
|    P101    | 2023-05-01 |     10     |    49.99   | North  |
|    P102    | 2023-05-02 |     8      |    24.99   | West   |
|    P103    | 2023-05-03 |     15     |    14.99   | South  |
|    P104    | 2023-05-04 |     5      |   199.99   | East   |
+------------+------------+------------+------------+--------+

# df_product_reviews
+------------+-------------+--------+-------------+
| product_id | review_date | rating | reviewer_id |
+------------+-------------+--------+-------------+
|    P101    |  2023-05-03 |   5    |    R501     |
|    P102    |  2023-05-04 |   3    |    R502     |
|    P103    |  2023-05-05 |   4    |    R503     |
|    P105    |  2023-05-06 |   2    |    R504     |
+------------+-------------+--------+-------------+

### Questions

1. For each product in the df_product_sales DataFrame, what is the total sales (units_sold multiplied by sale_price)
and average rating? Output a DataFrame showing product_id, total_revenue, and average_rating of each product. Assume
that some products in the df_product_sales may not have any reviews. For such products, impute the average_rating
with-99.

2. For the product_sales table, compute the day-on-day growth in total_revenue. Display the results in a new column
named DoD_growth. Remove the first day in the output DataFrame, and sort the rows based on the ascending order of date.

3. For the product_sales table, group by the region column and determine the product with the highest sales
(in terms of revenue) for each region. Assume that more than one product could have the same revenue values
for each region. Output the region, product_id, total_revenue.

"""

### Solution ###



# Setup
df_product_sales = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Python Pandas/Data/Amazon/product_sales.csv')
df_product_reviews = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Python Pandas/Data/Amazon/product_reviews.csv')


# 2. For the product_sales table, compute the day-on-day growth in total_revenue. Display the results in a new column
# named DoD_growth. Remove the first day in the output DataFrame, and sort the rows based on the ascending order of date.


# 3. For the product_sales table, group by the region column and determine the product with the highest sales
# (in terms of revenue) for each region. Assume that more than one product could have the same revenue values
# for each region. Output the region, product_id, total_revenue.
```

```python
df_product_sales
```

```python
df_product_reviews
```

```python
# 1. For each product in the df_product_sales DataFrame, what is the total sales (units_sold multiplied by sale_price)
# and average rating? Output a DataFrame showing product_id, total_revenue, and average_rating of each product. Assume
# that some products in the df_product_sales may not have any reviews. For such products, impute the average_rating
# with-99.

df_product_sales['total_revenue']= df_product_sales['units_sold']* df_product_sales['sale_price']
average_rating = df_product_reviews.groupby('product_id')['rating'].mean().reset_index()
average_rating.rename(columns={'rating':'average_rating'}, inplace=True)
```

```python
df_product_sales
```

```python
result = pd.merge(df_product_sales,average_rating, on='product_id',how='left')
```

```python
result
```

```python
result['average_rating'].fillna(-99)
```

```python
result[['product_id','total_revenue','average_rating']]
```

2. For the product_sales table, compute the day-on-day growth in total_revenue. Display the results in a new column
named DoD_growth. Remove the first day in the output DataFrame, and sort the rows based on the ascending order of date.

```python
df_product_sales = df_product_sales.sort_values(by='sale_date')
```

```python
df_product_sales
```

```python
df_product_sales['prev_date']= df_product_sales['total_revenue'].shift()
```

```python
df_product_sales
```

```python
df_product_sales['DoD_growth'] = df_product_sales['total_revenue']/df_product_sales['prev_date'] -1
```

```python
df_product_sales['DoD_growth']
```

```python
df_output = df_product_sales.loc[1:, ['sale_date', 'DoD_growth']]
```

```python
df_output
```

3. For the product_sales table, group by the region column and determine the product with the highest sales (in terms of revenue) for each region. Assume that more than one product could have the same revenue values
for each region. Output the region, product_id, total_revenue.

```python
df_product_sales['revenue']= df_product_sales['units_sold']* df_product_sales['sale_price']
df_product_sales['product_rank'] = df_product_sales.groupby('region')['revenue'].rank(method='dense', ascending=False);
#filter first rank
output_df=df_product_sales[df_product_sales['product_rank']==1]
output_df
```

```python
output_df.rename(columns={'revenue':'total_revenue'})
```

```python
output_df=output_df[['region', 'product_id', 'total_revenue']]
output_df
```
